```python
# Predicting Salary from Years of Experience
# Imagine we have data showing how salaries change with years of experience. We plot the data points and try to fit a line through them.
# Below is a Python graph to illustrate this:
# python
# CopyEdit
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data: Years of experience vs. Salary
X = np.array([1, 2, 3, 5, 7, 10, 15]).reshape(-1, 1)
y = np.array([30, 35, 40, 50, 65, 80, 120])

# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot the data
plt.scatter(X, y, color='blue', label="Actual Data")
plt.plot(X, y_pred, color='red', linewidth=2, label="Predicted Line")
plt.xlabel("Years of Experience")
plt.ylabel("Salary (in $1000s)")
plt.title("Simple ML Example: Experience vs. Salary")
plt.legend()
plt.show()
```

```python
# If we plot ice cream sales against temperature, we expect a line close to the data points, but not touching every point exactly.
# python
# CopyEdit
# Generate example data
np.random.seed(42)
X_temp = np.linspace(20, 40, 10).reshape(-1, 1)  # Temperature in Celsius
y_sales = 2 * X_temp.flatten() + np.random.normal(0, 5, len(X_temp))  # Sales with noise

# Fit a simple regression model
model = LinearRegression()
model.fit(X_temp, y_sales)
y_pred = model.predict(X_temp)

# Plot
plt.scatter(X_temp, y_sales, color='blue', label="Actual Data")
plt.plot(X_temp, y_pred, color='red', linewidth=2, label="Predicted Line")
plt.xlabel("Temperature (C)")
plt.ylabel("Ice Cream Sales")
plt.title("Balancing a Model with Randomness")
plt.legend()
plt.show()
```

```python
# 5. Bias-Variance Tradeoff
# Lets now visualize the problem of model complexity.
# 1. Too Simple: A Linear Line Through Quadratic Data
# 	If the data follows a quadratic pattern, but we fit a straight line, we miss important details (underfitting).
# python
# CopyEdit
X_quad = np.linspace(-5, 5, 100).reshape(-1, 1)
y_quad = X_quad.flatten()**2 + np.random.normal(0, 3, len(X_quad))  # Quadratic pattern

model.fit(X_quad, y_quad)
y_pred = model.predict(X_quad)

plt.scatter(X_quad, y_quad, color='blue', label="Actual Data")
plt.plot(X_quad, y_pred, color='red', linewidth=2, label="Linear Fit")
plt.title("Underfitting: A Too-Simple Model")
plt.legend()
plt.show()
```

```python
# 2. Too Complex: A Curve That Touches Every Point
# 	If we fit a high-degree polynomial to our data, it touches every point, but generalizes poorly (overfitting).
# CopyEdit
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Fit a high-degree polynomial model
poly_model = make_pipeline(PolynomialFeatures(degree=10), LinearRegression())
poly_model.fit(X_quad, y_quad)
y_poly_pred = poly_model.predict(X_quad)

plt.scatter(X_quad, y_quad, color='blue', label="Actual Data")
plt.plot(X_quad, y_poly_pred, color='red', linewidth=2, label="Overfitting Model")
plt.title("Overfitting: Too Complex Model")
plt.legend()
plt.show()
```
